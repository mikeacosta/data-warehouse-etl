#### Join all tables

```python
%%sql
SELECT to_timestamp(t.start_time,'YYYY-MM-DD HH24:MI:SS'), a.name, s.title, u.first_name, u.last_name
FROM songplays sp JOIN artists a ON a.artist_id = sp.artist_id
JOIN songs s ON s.song_id = sp.song_id
JOIN users u ON u.user_id = sp.user_id
JOIN time t ON t.start_time = sp.start_time
ORDER BY sp.start_time
```
